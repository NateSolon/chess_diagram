```python
def get_bbox(f):
    tf = str(f).replace('.png', '.txt')
    with open(tf, 'r') as file:
        contents = file.read()
    bbox = list(map(float, contents.split()[1:]))
    return bbox
```

```python
def draw_box(f):
    x,y,w,h = get_bbox(f)
    im = PILImage.create(f)
    imh, imw = im.shape

    x *= imw
    w *= imw
    y *= imh
    h *= imh

    left = round(x - w/2)
    right = round(x + w/2)
    top = round(y - h/2)
    bottom = round(y + h/2)

    corners = (left,top,right,bottom)
    di = ImageDraw.Draw(im)
    di.rectangle(corners, outline='red', width=5)
    return im
```

```python
def draw_box(im, box):
    x,y,w,h = box
    imh, imw = im.shape

    x *= imw
    w *= imw
    y *= imh
    h *= imh

    left = round(x - w/2)
    right = round(x + w/2)
    top = round(y - h/2)
    bottom = round(y + h/2)

    corners = (left,top,right,bottom)
    
    c = im.copy()
    di = ImageDraw.Draw(c)
    di.rectangle(corners, outline='red', width=5)
    
    return c
```
